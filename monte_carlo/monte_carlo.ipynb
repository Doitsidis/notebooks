```python

# This is a little messy, but the idea is clear enough.


import numpy as np
import pandas as pd
import matplotlib

# pip install empyrical ;)
from empyrical import max_drawdown, alpha_beta

import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from matplotlib.ticker import StrMethodFormatter, NullFormatter
```

```python

trades = pd.read_csv('./trades_subset.csv')

# 'gain_pc' is the only column we're interested in 
# expressed as a percentage in the csv file
gains = trades['gain'].divide(100) # reduce percentage to fractional

# Fraction of the account to trade with
size = 1

# Using the given trades in the csv, calculate cumulative product eq curve
empirical = np.cumprod(1 + (size * gains)) 

print( "Maximum drawdown: {0:.2f} %".format( max_drawdown( gains ) * 100 ) )
print( "Final account balance: {0:.2f} XBT".format( empirical.iloc[-1] ) )

```

```python

# Setup matplotlib styling

palette = plt.get_cmap('Set1')
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(18.5, 10.5)
plt.style.use('seaborn-darkgrid')

ax=plt.gca()
ax.set_yscale('log')

plt.yticks( np.arange(0, 50, step=2) )
ax.yaxis.set_major_formatter( StrMethodFormatter('{x:.1f}') )
ax.yaxis.set_minor_formatter( NullFormatter() )

font = { 'weight' : 'bold',
        'size'   : 12}

matplotlib.rc( 'font', **font )


# Use simple linear series
t = np.arange( 0, gains.size )



plt.title( "Simulated equity curves", loc='left', fontsize=22, fontweight=2, color='blue' )
plt.ylabel( "Balance (BTC)" )
plt.xlabel( "Trade #" )



# Plot the empirically observed eq curve given in the CSV
plt.plot(t, empirical, c='b')

# Save the list of drawdowns for each simulation for later
dds = []

# Re-run the below iteration each time reducing the `size` to find optimal balance fraction
size = 0.75
max_risk = 25

# Run n simulations, suggest 1000, using 100 for illustration: 
num_samples = 100 


# Perform the simulations
for i in range( num_samples ):
    # Sample with replace, bootstrap
    gainsr = pd.Series( np.random.choice(gains, gains.size, replace=True) )    
    
    # Calculate equity curve
    res = np.cumprod( 1 + (size * gainsr) ) 
    
    # Plot this simulation's cure
    plt.plot( t, res, c=np.random.rand(3,) )

    # Get its' max drawdown and save in a list for use later
    dds.append( max_drawdown( gainsr * size ) )


plt.show()
plt.clf()
```

```python


# List of all drawdowns, one for each of the simulations (num_samples)
ddss = pd.Series( -100 * np.sort( dds ) )

# Percentage of the drawdown samples that were within our risk tolerance, e.g. 95% 
pc_under_risk = ( ( ddss[ ddss <= max_risk ].count() )  / num_samples ) * 100



# Optimally, this should be 95% or more (according to risk appetite) -- keep iterating until it is
# This is the important bit that determines risk
print( "Drawdowns under max_risk: {0:.2f} %".format( pc_under_risk ) )


```

```python
# Numerical array representing 0.5% percentage width bins to arrange the histo data into
bins = np.arange(0, 101, step=0.5)

# Sort the data into the bins above
dd_bins = pd.cut( ddss, bins )

# Show the values for each bin
# print(dd_bins.value_counts(normalize=True))

plt.title("Drawdown histogram (PMF)", loc='left', fontsize=22, fontweight=2, color='blue')


font = {'weight' : 'bold',
        'size'   : 14}

matplotlib.rc('font', **font)

plt.xlabel( "Drawdown (in 0.5% wide bins)", fontsize=18 )
plt.ylabel( "% of simulations with the given drawdown bin", fontsize=18 )
ax = ( dd_bins.value_counts(sort=False, normalize=True) * 100).plot.bar( rot=0, color="b", figsize=(20,14) )

# Hide messy matplotlib ticks, cant figure this bit out properly sorry!
ax.axes.get_xaxis().set_ticks([])

plt.show()

```

```python
plt.clf()
plt.title("Inverse cumulative dist. function of maximum drawdown", loc='left', fontsize=22, fontweight=2, color='blue')

# Running sum of probabilities for each drawdown 'band'
norms = (dd_bins.value_counts(sort=False,normalize=True) * 100).cumsum()

fig = matplotlib.pyplot.gcf()
fig.set_size_inches(18.5, 10.5)

plt.ylabel("Target maximum drawdown %", fontsize=18)
plt.xlabel("probability of limiting drawdown < maximum target ", fontsize=18)

# Crop the data to chop tail show most relevent section
ax = plt.plot(norms[:100]/100, np.arange(0, 100.5, step=0.5)[:100])

plt.show()

```
